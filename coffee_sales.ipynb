```python
import pandas as pd

# Load the CSV file into a DataFrame
coffee = pd.read_csv('coffee_sales.csv')

# Display the first 5 rows
coffee.head()
```

```python
display(coffee.info())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example: aggregate sales by coffee_name
product_df = coffee.groupby('coffee_name').agg({'money':'sum'}).reset_index()

# Bar plot: Total sales by coffee_name
plt.figure(figsize=(10,6))
sns.barplot(x='coffee_name', y='money', data=product_df, palette='viridis')
plt.title('Total Sales by Coffee Type')
plt.xlabel('Coffee Name')
plt.ylabel('Total Sales (Money)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Aggregate sales by hour_of_day
hour_df = coffee.groupby('hour_of_day').agg({'money':'sum'}).reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(x='hour_of_day', y='money', data=hour_df, marker='o')
plt.title('Total Sales by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales (Money)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Aggregate sales by weekday
weekday_df = coffee.groupby('Weekday').agg({'money':'sum'}).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Weekday', y='money', data=weekday_df, palette='mako')
plt.title('Total Sales by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Total Sales (Money)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Total sales by coffee_name and hour_of_day
product_df = coffee.groupby(['coffee_name', 'hour_of_day']).agg({'money':'sum'}).reset_index()
plt.figure(figsize=(10,6))
sns.scatterplot(x='hour_of_day', y='money', size='money', hue='coffee_name',
                data=product_df, palette='coolwarm', sizes=(50,300), legend='brief')
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.title('Total Sales by Coffee Type and Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales (Money)')
plt.tight_layout()
plt.show()
```
